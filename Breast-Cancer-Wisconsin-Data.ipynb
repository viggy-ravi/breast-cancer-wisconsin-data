# Breast Cancer Wisconsin Data
Source: https://www.kaggle.com/uciml/breast-cancer-wisconsin-data

Attribute Information:

1) ID number

2) Diagnosis (M = malignant, B = benign)

3) Cells 3-32: Ten real-valued features are computed for each cell nucleus:

    a) radius (mean of distances from center to points on the perimeter)
    b) texture (standard deviation of gray-scale values)
    c) perimeter
    d) area
    e) smoothness (local variation in radius lengths)
    f) compactness (perimeter^2 / area - 1.0)
    g) concavity (severity of concave portions of the contour)
    h) concave points (number of concave portions of the contour)
    i) symmetry
    j) fractal dimension ("coastline approximation" - 1)

The mean, standard error and "worst" or largest (mean of the three
largest values) of these features were computed for each image,
resulting in 30 features. For instance, field 3 is Mean Radius, field
13 is Radius SE, field 23 is Worst Radius.

All feature values are recoded with four significant digits.

Missing attribute values: none

Class distribution: 357 benign, 212 malignant

# Prediction Algorithms

```python
import time
from sklearn import neighbors, svm
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
import operator
```

```python
# Split into X and y datasets
X = np.array(df_norm.drop(['diagnosis'],axis=1))
y = np.array(df_norm['diagnosis'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## kNN

```python
# kNN
start_time = time.time()
model_kNN = neighbors.KNeighborsClassifier(3, weights='distance')
model_kNN.fit(X_train,y_train)
accuracy = model_kNN.score(X_test, y_test)
y_pred_kNN = model_kNN.predict(X_test)

print("acc: %s" %(round(accuracy, 4)))
print("---- %s seconds" % (round(time.time() - start_time, 4)))
```

```python
# Find best k value
kNN_acc = []
k_vals = []
for k in range(1,20,2):
    model_kNN = neighbors.KNeighborsClassifier(k, weights='distance')
    model_kNN.fit(X_train,y_train)
    accuracy = model_kNN.score(X_test, y_test)
    kNN_acc.append(accuracy)
    k_vals.append(k)

index, value = max(enumerate(kNN_acc), key=operator.itemgetter(1))
print("Found best k value to be %s with accuracy %s" %(k_vals[index], round(value,4)))
```

## SVM

```python
# SVM
## kernel = linear
## C = regularization parameter
## gamma = scale (default), auto -- for RBF, poly, sigmoid

start_time = time.time()
model_SVM = svm.SVC(kernel='linear', C=1)
model_SVM.fit(X_train,y_train)
accuracy = model_SVM.score(X_test, y_test)
y_pred_SVM = model_SVM.predict(X_test)

print("acc: %s" %(round(accuracy, 4)))
print("---- %s seconds" % (round(time.time() - start_time, 4)))
```

```python
# Find best regularization parameter value (C)
SVM_acc = []
reg_param = []
for c in np.arange(1,4,0.1):
    model_SVM = svm.SVC(kernel='linear', C=c)
    model_SVM.fit(X_train,y_train)
    accuracy = model_SVM.score(X_test, y_test)
    SVM_acc.append(accuracy)
    reg_param.append(c)
    
index, value = max(enumerate(SVM_acc), key=operator.itemgetter(1))
print("Found best C to be %s with accuracy %s" %(reg_param[index], round(value,4)))
```

## LDA

```python
# LDA
start_time = time.time()
model_LDA = LDA()
model_LDA.fit(X_train,y_train)
accuracy = model_LDA.score(X_test, y_test)
y_pred_LDA = model_LDA.predict(X_test)

print("acc: %s" %(round(accuracy, 4)))
print("---- %s seconds" % (round(time.time() - start_time, 4)))
```

## Nerual Network

```python
# NN
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# For neural networks, y needs to be a binary class matrix
yy = to_categorical([0 if x=='B' else 1 for x in y])
X_train_NN, X_test_NN, y_train_NN, y_test_NN = train_test_split(X, yy, test_size=0.2, random_state=8)
```

```python
# Create neural network model (3 layers)
# Input layer will take in all 30 parameters
# Output layer will output either benign (0) or malignant (1)

model = keras.Sequential()
model.add(Dense(30, activation="relu"))
model.add(Dense(16, activation="relu"))
model.add(Dense(2, activation="softmax"))
```

```python
model.compile(loss='categorical_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])

model.fit(X_train_NN, y_train_NN, epochs=100, batch_size=32, verbose=2)

model.evaluate(X_test_NN, y_test_NN)
```

```python
y_pred = np.argmax(model.predict(X_test_NN), axis=1)
y_true = np.argmax(y_test_NN, axis=1)

matrix = confusion_matrix(y_true, y_pred)
report = classification_report(y_true, y_pred, digits=3)
print(report)

df_cm = pd.DataFrame(matrix)
plt.figure(figsize = (10,7))
sns.heatmap(df_cm, cmap="YlGnBu", annot=True, linewidths=.1)
```

# Dimensionality Reduction

## Dimensionality Reduction - PCA

```python
# PCA 
model_PCA = PCA(n_components=6)
X_PCA = model_PCA.fit_transform(X)

print(model_PCA.explained_variance_ratio_)
# print(X_PCA)
```

```python
# Re-split data with new PCA inputs
X_train_PCA, X_test_PCA, y_train_PCA, y_test_PCA = train_test_split(X_PCA, y, test_size=0.2)
```

```python
# PCA+kNN
start_time = time.time()
model_kNN = neighbors.KNeighborsClassifier(15, weights='distance')
model_kNN.fit(X_train_PCA,y_train_PCA)
accuracy = model_kNN.score(X_test_PCA, y_test_PCA)
y_pred_kNN = model_kNN.predict(X_test_PCA)

print("acc: %s" %(round(accuracy, 4)))
print("---- %s seconds" % (round(time.time() - start_time, 4)))
```

```python
# Find best k value
kNN_acc = []
k_vals = []
for k in range(1,100,2):
    model_kNN = neighbors.KNeighborsClassifier(k, weights='distance')
    model_kNN.fit(X_train_PCA,y_train_PCA)
    accuracy = model_kNN.score(X_test_PCA, y_test_PCA)
    kNN_acc.append(accuracy)
    k_vals.append(k)
    
index, value = max(enumerate(kNN_acc), key=operator.itemgetter(1))
print("Found best k value to be %s with accuracy %s" %(k_vals[index], round(value,4)))
```

```python
# PCA+SVM
start_time = time.time()
model_SVM = svm.SVC(kernel='linear', C=1.0)
model_SVM.fit(X_train_PCA,y_train_PCA)
accuracy = model_SVM.score(X_test_PCA, y_test_PCA)
y_pred_SVM = model_SVM.predict(X_test_PCA)

print("acc: %s" %(round(accuracy, 4)))
print("---- %s seconds" % (round(time.time() - start_time, 4)))
```
